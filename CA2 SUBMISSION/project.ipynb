```python
import tkinter as tk
from tkinter import messagebox

# Creating the menu dictionary
menu = {
        "Jollof Rice": 350,
        "Coconut Fried Rice": 350,
        "Jollof Spaghetti": 350,
        "Sweet Chilli Chicken": 300,
        "Grilled Chicken Wings": 400,
        "Fried Beef": 400,
        "Fried Fish": 500,
        "Boiled Egg": 200,
        "Sauteed Sausages": 200,
        "Savoury Beans": 350,
        "Roasted Sweet Potato": 300,
        "Fried Plantains": 150,
        "Mixed Vegetable Salad": 150,
        "Boiled Yam": 150,
        "Eba": 100,
        "Poundo Yam": 100,
        "Semo": 100,
        "Atama Soup": 450,
        "Egusi Soup": 480,
        "Sweet Chilli Chicken": 1100,
        "Grilled Chicken Wings": 400,
        "Fried Beef": 400,
        "Fried Fish": 500,
        "Boiled Egg": 200,
        "Sauteed Sausages": 200,
        "Water": 200,
        "Glass Drink(35cl)": 150,
        "PET Drink(35cl)": 300,
        "PET Drink(30cl)": 350,
        "Glass/Canned Malt": 500,
        "Fresh Yo": 600,
        "Pineapple Juice": 350,
        "Mango Juice": 350,
        "Zobo Drink": 350,
    } 
# displaying the menu  
def display_menu():
    menu_text = """
WELCOME TO PAU CAFETARIA

PLEASE SCROLL TO VIEW OTHER VARITIES ON THE MENU
PLEASE ENTER EACH FONT AS SEEN ON THE MENU
           RICE/PASTA                           
    Jollof Rice          350              
    Coconut Fried Rice   350           
    Jollof Spaghetti     350             
                                          
           SIDE DISHES                    
    Savoury Beans            350          
    Roasted Sweet Potatoes   300
    Fried Plantains          150
    Mixed Vegetable Salad    150                 
    Boiled Yam               150          
                                          
        SOUPS & SWALLOWS                  
    Eba               100                 
    Poundo Yam        100               
    Semo              100                 
    Atama Soup        450               
    Egusi Soup        480              
    
       PROTEINS
    Sweet Chilli Chicken    1100
    Grilled Chicken Wings    400
    Fried Beef               400
    Fried Fish               500
    Boiled Egg               200
    Sauteed Sausages         200

        BEVERAGES
     Water                   200
     Glass Drink(35cl)       150
     PET Drink(35cl)         300
     PET Drink(30cl)         350
     Glass/Canned Malt       500 
     Fresh Yo                600
     Pineapple Juice         350
     Mango Juice             350
     Zobo Drink              350
       THANKS! AND ENJOY!
    """
    menu_text_widget.delete(1.0, tk.END)
    menu_text_widget.insert(tk.END, menu_text)
    menu_text_widget.config(bg="black", fg="white")

# defining the menu contents
def submit():
    name = name_entry.get()
    riceorpasta = riceorpasta_entry.get()
    portions = int(portions_entry.get())
    proteins = proteins_entry.get()
    quantity = int(quantity_entry.get())
    sidedishes = sidedishes_entry.get()
    portions1 = int(portions1_entry.get())
    beverages = beverages_entry.get()
    bottles = int(bottles_entry.get())
    soupsandswallows = soupsandswallows_entry.get()
    portions2 = int(portions2_entry.get())

    chosen_items_total = 0
    chosen_items_with_quantities = [(riceorpasta, portions),
                                    (proteins, quantity),
                                    (sidedishes, portions1),
                                    (beverages, bottles),
                                    (soupsandswallows, portions2)]
    
    for item, quantity in chosen_items_with_quantities:
        price_per_meal = menu.get(item,0 )
        if price_per_meal > 0:
            chosen_items_total += price_per_meal * quantity
        else:
            messagebox.showerror("Error", f"{item} is not available in the menu")

    # Determine the discount based on the total charge
    discount = 0
    if chosen_items_total < 2500:
        discount = 0.1
    elif 2500 <= chosen_items_total < 5000:
        discount = 0.15
    elif chosen_items_total >= 5000:
        discount = 0.25
    else:
         messagebox.showerror("Error", "information cannot be processed")

    # Calculate the discounted total
    discounted_total = chosen_items_total - (chosen_items_total * discount)
    
    # Display the total charge to the user
    messagebox.showinfo("Total Charge", f"Total Charge: {chosen_items_total}\n Discount Price: {discounted_total}")
    
   
    # Display order information
    summary_text = f"Customer Name: {name}\nRice or Pasta: {riceorpasta}\nProteins: {proteins}\nSide dishes: {sidedishes}\nBeverages: {beverages}\nSoups and Swallows: {soupsandswallows}"
    messagebox.showinfo("Order Information", summary_text)


# Create main window
root = tk.Tk()
root.title("Menu")
root.geometry("950x400")

# Create a text widget to display the menu
menu_text_widget = tk.Text(root, height=10, width=40)
menu_text_widget.pack(pady=10)

# Button to trigger displaying the menu
display_button = tk.Button(root, text="Display Menu", bg="blue", fg="black", command=display_menu)
display_button.pack(pady=5)
    
# Create name label and entry
name_label = tk.Label(root, text="Name:")
name_label.pack()
name_entry = tk.Entry(root)
name_entry.pack()

# Create riceorpasta label and entry
riceorpasta_label = tk.Label(root, text="Rice or Pasta:")
riceorpasta_label.pack()
riceorpasta_entry = tk.Entry(root)
riceorpasta_entry.pack()

# Create portions label and entry
portions_label = tk.Label(root, text="No. of portions:")
portions_label.pack()
portions_entry = tk.Entry(root)
portions_entry.pack()

# Create proteins label and entry
proteins_label = tk.Label(root, text="Proteins:")
proteins_label.pack()
proteins_entry = tk.Entry(root)
proteins_entry.pack()

# Create quantity label and menu
quantity_label = tk.Label(root, text="Quantity:")
quantity_label.pack()
quantity_entry = tk.Entry(root)
quantity_entry.pack()

# Creatw sidedishes label and menu
sidedishes_label = tk.Label(root, text="Side dishes:")
sidedishes_label.pack()
sidedishes_entry = tk.Entry(root)
sidedishes_entry.pack()

# Create portions1 label and menu
portions1_label = tk.Label(root, text="Portions:")
portions1_label.pack()
portions1_entry = tk.Entry(root)
portions1_entry.pack()

# Create beverages label and menu
beverages_label = tk.Label(root, text="Beverages:")
beverages_label.pack()
beverages_entry = tk.Entry(root)
beverages_entry.pack()

# Create bottles label and menu
bottles_label = tk.Label(root, text="Bottles:")
bottles_label.pack()
bottles_entry = tk.Entry(root)
bottles_entry.pack()

# Create soupsandswallows label and menu
soupsandswallows_label = tk.Label(root, text="Soups and Swallows:")
soupsandswallows_label.pack()
soupsandswallows_entry = tk.Entry(root)
soupsandswallows_entry.pack()

# Create portions2 label and menu
portions2_label = tk.Label(root, text="Portions:")
portions2_label.pack()
portions2_entry = tk.Entry(root)
portions2_entry.pack()


# Create submit button
submit_button = tk.Button(root, text="Place Order", bg="blue", fg="black",command=submit)
submit_button.pack()

# Run the main event loop
root.mainloop()
```
